# The Monty Hall Problem with 4 Choices — Simulation and Analysis

I once heard a story about a probability puzzle: if you have 4 choices and pick one, then the host cancels one choice that is neither your choice nor the correct one, switching your choice supposedly increases your chances of winning.

I couldnot visualize how that works, so I decided to simulate it and calculate whether this claim is true or false.

```python
import random
```

# Case 1: Staying with your original choice
# In this part, we simulate the situation where you do not change your initial choice.
# We expect the probability of winning to be 1 out of 4, or 25%.

---



```python
counter_stayed = 0
for i in range(1000000):
  real_door = random.randint(1,4)
  your_choise = random.randint(1,4)
  if real_door == your_choise:
    counter_stayed += 1
print("the percentage of your choice being correct is " + str(counter_stayed*100/1000000) + "%")
```

As we can observe, the result is approximately 25%, which makes sense since there is only one correct choice out of four. Therefore, if you always stay with your first pick, your chance of winning remains about 25%.

---





# Case 2: Switching to another choice
# In this scenario, the host removes one incorrect option that is neither your choice nor the correct one.
# Then, you switch to a different remaining option.
# We will now measure how often this strategy results in winning.

```python
counter_changed = 0
for i in range(1000000):
  real_door = random.randint(1,4)
  your_choise = random.randint(1,4)
  removed_choise = random.randint(1,4)
  while (removed_choise == your_choise) or (removed_choise == real_door):
    removed_choise = random.randint(1,4)

  #swapping my choise
  new_choise = random.randint(1,4)
  while (new_choise == your_choise) or (new_choise == removed_choise):
    new_choise = random.randint(1,4)

  if real_door == new_choise:
    counter_changed += 1
print("the percentage of your choice being correct is " + str(counter_changed*100/1000000) + "%")
```

As we can see, the probability increases to around 37.5%.
This confirms that the so-called "Monty Hall effect" is real for four doors —
switching your choice after one wrong option is revealed actually increases your winning chance.
